```python
import matplotlib.pyplot as plt
import pandas as pd

# Function to read polygon points from a CSV file
def read_polygon_from_csv(file_path):
    data = pd.read_csv(file_path, header=None)
    x = data[0].values
    y = data[1].values
    return x, y

# File paths for the two CSV files
file1 = "/home/fleeting/ISU/sem1/COMS_5760/project/UAVSurveillance/src/ground_control/config/path_polygon.csv"  # Replace with your first file's path
file2 = "/home/fleeting/ISU/sem1/COMS_5760/project/UAVSurveillance/src/ground_control/config/shape_polygon.csv"  # Replace with your second file's path

# Read the polygon points
x1, y1 = read_polygon_from_csv(file1)
x2, y2 = read_polygon_from_csv(file2)

# Plot the polygons
plt.figure(figsize=(8, 8))
plt.plot(x1, y1, label="Polygon 1", color="red", linewidth=2)  # First polygon in blue
plt.plot(x2, y2, label="Polygon 2", color="blue", linewidth=2)   # Second polygon in red

# Adding details to the plot
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Polygon Boundaries")
plt.legend()
plt.grid(True)
plt.axis('equal')  # Ensure equal scaling for x and y axes

# Show the plot
plt.show()
```
